```python
# import dependencies
import pandas as pd
import numpy as np
from citipy import citipy
import requests

# import OpenWeatherMap API Key
import sys
sys.path.append('../')
from config import weather_api_key
```

```python
# 3: generate random latitudes and longitudes for city coordinates
lats = np.random.uniform(low = -90, high = 90, size = 2000)
lngs = np.random.uniform(low = -180, high = 180, size = 2000)
coordinates = list(zip(lats, lngs))
```

```python
# 4: get nearest city names
cities = []

# get the nearest city for all coordinates
for coord in coordinates:
    city = citipy.nearest_city(coord[0], coord[1]).city_name
    
    # do not allow duplicates
    if city not in cities:
        cities.append(city)
```

```python
# Set up URL for Weather Map API Call.
api_url = 'http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=' + weather_api_key

# retrieve city data
city_data = []
for city in cities:
    city_url = api_url + '&q=' + city.replace(' ', '+')
    
    try:
        # 5: get city weather data from OpenWeatherMap
        city_weather = requests.get(city_url).json()
        city_lat = round(city_weather["coord"]["lat"], 2) # rounding to match instructions image
        city_lng = round(city_weather["coord"]["lon"], 2) # rounding to match instructions image
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        
        # get all current weather by looping through all weather IDs and combining the descirptions
        city_current = ''
        for weather in city_weather["weather"]:
            if city_current == '': # first desc, don't add a space
                city_current += weather['description']
            else: # after first desc, add a comma and a space
                city_current += ', ' + weather['description']
           
        # 6: add fecthed data to city_data
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_current})
        
    # city not found
    except:
        print("City not found. Skipping...")
        pass
```

```python
# 7: convert to a dataframe
city_data_df = pd.DataFrame(city_data)

# 8: export data to weather_database folder
city_data_df.to_csv('WeatherPy_Database.csv')
```
